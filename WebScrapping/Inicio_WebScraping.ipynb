# Iniciando WebScraping

```python
bloco_infos = soup.find_all("div", class_="product-box-detail")

bloco_infos[0]
```

```python
todos_produtos = []
NUM_PAGINAS = 100 

#essa função auxilia caso .find retorne None, já que as vezes podem ter produtos que nao possuem todas as classes, aí não da erro na extração
def webscraping(produto, parametro, chave):
        alvo = produto.find(parametro, class_=chave)
        return alvo.get_text().strip() if alvo else "N/A"
    
for pagina in range(1, NUM_PAGINAS + 1):
    url = f"https://www.dafiti.com.br/calcados-femininos/tamanho-35/?seller=Dafiti&page={pagina}"
    response = requests.get(url, headers=headers)
    print(f"Página {pagina}: {response.status_code}")
    
    soup = BeautifulSoup(response.text, "html.parser")
    bloco_infos = soup.find_all("div", class_="product-box-detail")
    todas_entradas =[]
    for bloco in bloco_infos:
        marca = webscraping(bloco, "div", "product-box-brand")
        cor = webscraping(bloco, "div", "product-box-colors")
        vendedor = webscraping(bloco, "div", "product-box-seller")
        produto = webscraping(bloco, "p", "product-box-title")
        tags = webscraping(bloco, "span", "product-box-category-name")
        preco_orig = webscraping(bloco, "span", "product-box-price-from")
        preco_pix = webscraping(bloco, "span", "product-box-price-to")
        preco_parcelado = webscraping(bloco, "span", "product-box-installment-value")
        qtd_parcela = webscraping(bloco, "span", "product-box-installment-parcel")
    
        entrada = {
            "Marca": marca, "Cor": cor, "Vendedor": vendedor,
            "Produto": produto, "Tags": tags, "Preço Original": preco_orig,
            "Preço Pix": preco_pix, "Preço Parcelado": preco_parcelado,
            "Quantidade de Parcelas": qtd_parcela
        }
        todas_entradas.append(entrada)
    todos_produtos.extend(todas_entradas)

print(f"\nTotal coletado: {len(todos_produtos)}")
```

```python
dados = pd.DataFrame(todos_produtos)
```

```python
dados.to_csv('20260412_DadosDafiti.csv', index=False)
```
