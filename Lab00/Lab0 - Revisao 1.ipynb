## Passo 1 – Criar o dicionário principal

```python
distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": {"media": None, "med": None, "var": None, "dp": None}},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": {"media": None, "med": None, "var": None, "dp": None}},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": {"media": None, "med": None, "var": None, "dp": None}}
}
```

## Passo 2 – Importar bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Passo 3 – Gerar os dados

```python
distribuicoes["Normal"]["dados"] = np.random.normal(0, 1, size=1000)
distribuicoes["Exponencial"]["dados"] = np.random.exponential(2, size=1000)
distribuicoes["Uniforme"]["dados"] = np.random.uniform(-1, 1, size=1000)
```

## Passo 4 – Calcular estatísticas básicas

```python
distribuicoes["Normal"]["estatisticas"]["media"] = np.mean(distribuicoes["Normal"]["dados"])
distribuicoes["Exponencial"]["estatisticas"]["media"] = np.mean(distribuicoes["Exponencial"]["dados"])
distribuicoes["Uniforme"]["estatisticas"]["media"] = np.mean(distribuicoes["Uniforme"]["dados"])
```

```python
distribuicoes["Normal"]["estatisticas"]["med"] = np.median(distribuicoes["Normal"]["dados"])
distribuicoes["Exponencial"]["estatisticas"]["med"] = np.median(distribuicoes["Exponencial"]["dados"])
distribuicoes["Uniforme"]["estatisticas"]["med"] = np.median(distribuicoes["Uniforme"]["dados"])
```

```python
distribuicoes["Normal"]["estatisticas"]["var"] = np.var(distribuicoes["Normal"]["dados"])
distribuicoes["Exponencial"]["estatisticas"]["var"] = np.var(distribuicoes["Exponencial"]["dados"])
distribuicoes["Uniforme"]["estatisticas"]["var"] = np.var(distribuicoes["Uniforme"]["dados"])
```

```python
distribuicoes["Normal"]["estatisticas"]["dp"] = np.std(distribuicoes["Normal"]["dados"])
distribuicoes["Exponencial"]["estatisticas"]["dp"] = np.std(distribuicoes["Exponencial"]["dados"])
distribuicoes["Uniforme"]["estatisticas"]["dp"] = np.std(distribuicoes["Uniforme"]["dados"])
```

## Passo 5 – Visualizar os dados

```python
plt.hist(distribuicoes["Normal"]["dados"])
```

```python
plt.hist(distribuicoes["Exponencial"]["dados"])
```

```python
plt.hist(distribuicoes["Uniforme"]["dados"])
```

## Passo 6 – Organizar resultados em Pandas

```python
pd.DataFrame({distribuicao: distribuicoes[distribuicao]["estatisticas"] for distribuicao in distribuicoes.keys()}).T
```

## Passo 7 – Adicionar novas distribuições

```python
distribuicoes.update({"Poisson": {"params": {"scale": 1}, "dados": np.random.poisson(1, size=1000), "estatisticas": {"media": None, "med": None, "var": None, "dp": None}}})
distribuicoes["Poisson"].update({"estatisticas": {"media": np.mean(distribuicoes["Poisson"]["dados"]),
                                                 "med": np.median(distribuicoes["Poisson"]["dados"]),
                                                 "var": np.var(distribuicoes["Poisson"]["dados"]),
                                                 "dp": np.std(distribuicoes["Poisson"]["dados"])}})
```

## Passo 8 – Estatísticas adicionais

```python
distribuicoes["Normal"]["estatisticas"]["min"] = np.min(distribuicoes["Normal"]["dados"])
distribuicoes["Exponencial"]["estatisticas"]["min"] = np.min(distribuicoes["Exponencial"]["dados"])
distribuicoes["Uniforme"]["estatisticas"]["min"] = np.min(distribuicoes["Uniforme"]["dados"])
distribuicoes["Poisson"]["estatisticas"]["min"] = np.min(distribuicoes["Poisson"]["dados"])
```

```python
distribuicoes["Normal"]["estatisticas"]["max"] = np.max(distribuicoes["Normal"]["dados"])
distribuicoes["Exponencial"]["estatisticas"]["max"] = np.max(distribuicoes["Exponencial"]["dados"])
distribuicoes["Uniforme"]["estatisticas"]["max"] = np.max(distribuicoes["Uniforme"]["dados"])
distribuicoes["Poisson"]["estatisticas"]["max"] = np.max(distribuicoes["Poisson"]["dados"])
```

```python
q = [0, 25, 50, 75, 100]

distribuicoes["Normal"]["estatisticas"]["quartis"] = np.percentile(distribuicoes["Normal"]["dados"], q)
distribuicoes["Exponencial"]["estatisticas"]["quartis"] = np.percentile(distribuicoes["Exponencial"]["dados"], q)
distribuicoes["Uniforme"]["estatisticas"]["quartis"] = np.percentile(distribuicoes["Uniforme"]["dados"], q)
distribuicoes["Poisson"]["estatisticas"]["quartis"] = np.percentile(distribuicoes["Poisson"]["dados"], q)
```

```python
pd.DataFrame({distribuicao: distribuicoes[distribuicao]["estatisticas"] for distribuicao in distribuicoes.keys()}).T
```

## Passo 9 – Comparar distribuições graficamente

```python
dados = [
    distribuicoes["Normal"]["dados"],
    distribuicoes["Exponencial"]["dados"],
    distribuicoes["Uniforme"]["dados"],
    distribuicoes["Poisson"]["dados"],
]

plt.boxplot(dados, labels=["Normal", "Exponencial", "Uniforme", "Poisson"])
plt.show()
```

## Passo 10 – Comparar estatísticas calculadas com valores teóricos

```python
for nome, dist in distribuicoes.items():
    params = dist["params"]
    
    if nome == "Normal":
        media_teorica = params["loc"]
        
    elif nome == "Exponencial":
        media_teorica = params["scale"]
        
    elif nome == "Uniforme":
        a = params["low"]
        b = params["high"]
        media_teorica = (a + b) / 2

    elif nome == "Poisson":
        media_teorica = params["scale"]
    else:
        continue

    media_amostra = dist["estatisticas"]["media"]

    plt.figure()
    
    valores = [media_teorica, media_amostra]
    labels = ["Teórica", "Amostral"]
    
    plt.bar(labels, valores)
    plt.title(f"Distribuição {nome}")
    plt.ylabel("Média")
    
    plt.show()
```
