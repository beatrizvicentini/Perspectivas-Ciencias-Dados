## Exercício 3
Implemente uma função que recebe dois vetores NumPy e retorna:  
- Produto escalar (`np.dot`).  
- Distância euclidiana entre eles (`np.linalg.norm`).  

```python
def funcao(vetor1, vetor2):
    prod_esc = np.dot(vetor1, vetor2)
    dist = np.linalg.norm(vetor1 - vetor2)
    return prod_esc, dist
```

## Exercício 4
Crie uma matriz quadrada $A \in \mathbb{R}^{4 \times 4}$ com valores inteiros aleatórios entre $-5$ e $5$.  
- Verifique se $A$ é invertível calculando $\det(A)$ (`np.linalg.det`).  
- Caso seja, calcule $A^{-1}$ (`np.linalg.inv`) e verifique se $A \cdot A^{-1} = I$.  

```python
A = np.random.randint(-5, 6, (4, 4))
det_A = np.linalg.det(A)

if (det_A) != 0:
    A_inv = np.linalg.inv(A)
    prod_esc = np.dot(A, A_inv)
    print(prod_esc)
else:
    print("Não invertível.")

## Se aproxima da matriz identidade 4x4
```

## Exercício 5
Gere uma matriz simétrica $B \in \mathbb{R}^{5 \times 5}$ a partir de uma matriz aleatória $M$ definida por $B = \tfrac{1}{2}(M + M^T)$.  
- Calcule os autovalores e autovetores (`np.linalg.eig`).  
- Verifique numericamente a relação $Bv = \lambda v$.  

```python
M = np.random.rand(5, 5)
B = (M + M.T)
autovalores, autovetores = np.linalg.eig(B)

v1 = autovetores[:, 0]
lam1 = autovalores[0]
erro1 = np.linalg.norm(np.dot(B, v1) - lam1 * v1)
print(erro1)

v2 = autovetores[:, 1]
lam2 = autovalores[1]
erro2 = np.linalg.norm(np.dot(B, v2) - lam2 * v2)
print(erro2)

v3 = autovetores[:, 2]
lam3 = autovalores[2]
erro3 = np.linalg.norm(np.dot(B, v3) - lam3 * v3)
print(erro3)

v4 = autovetores[:, 3]
lam4 = autovalores[3]
erro4 = np.linalg.norm(np.dot(B, v4) - lam4 * v4)
print(erro4)

v5 = autovetores[:, 4]
lam5 = autovalores[4]
erro5 = np.linalg.norm(np.dot(B, v5) - lam5 * v5)
print(erro5)


## Erro muito pequenos
```

## Exercício 6
Considere um sistema linear $Ax = y$, com $A \in \mathbb{R}^{3 \times 3}$ e $y \in \mathbb{R}^3$.  
- Gere $A$ de forma aleatória e $y$ como um vetor de inteiros.  
- Resolva o sistema (`np.linalg.solve`).  
- Verifique a solução computando $\|Ax - y\|_2$ (`np.linalg.norm`).  

```python
A = np.random.rand(3, 3)
y = np.random.randint(1, 10, 3)
x = np.linalg.solve(A, y)

residuo = np.linalg.norm(np.dot(A, x) - y)
print(residuo)

## Resíduo epqueno
```

## Exercício 7
Simule 100 vetores bidimensionais com distribuição normal padrão e forme a matriz de covariância empírica $S$.  
- Calcule a matriz de covariância (`np.cov`).  
- Encontre os autovalores e autovetores de $S$ (`np.linalg.eig`).  
- Mostre que os autovetores correspondem às direções principais da variabilidade dos dados.

```python
dados = np.random.standard_normal((100, 2))
S = np.cov(dados, rowvar=False)
autovalores, autovetores = np.linalg.eig(S)

print(autovetores)
```
