# Comandos Úteis que Aprendemos na Aula
## Terminal

1. "cd": acessa diretórios pelo nome (.. volta na pasta anterior);
3. "ctrl + shft + v": cola links no terminal;
4. "ls": lista todos os arquivos contidos na pasta aberta que não estão coultos;
5. "ls -a": lista todos os arquivos contidos na pasta aberta, incluindo os coultos.


## Git

1. "git status": Indica o status do repositório na máquina em relação ao repositório na nuvem (arquivos divergentes que foram adicionados, excluídos ou alterados);
2. "git clone": Clona na máquina o repositório da nuvem indicado pelo link colado;
3. "git diff": Indica as divergências entre a main (repositório nuvem) e o repositório na máquina;
4. "git user.email": Configura o email que será identificado no commit;
5. "git user.name": Configura o nome que será identificado no commit;
6. "git add": adiciona alterações no commit (. adiciona tudo);
7. "git commit": sobe alterações para nuvem sem publicar;
8. "-m": mensagem opcional (porém recomendada) no commit para indicar as alterações que foram feitas;
9. "git push": envia as alterações para a main, se tornando a versão mais atualizada;
10. "git pull": puxa a versão mais atualizada do repositório a partir da nuvem, não sendo preciso clonar ele novamente.


## UV

1. "uv init": inicializa a ferramenta uv para rodar códigos python;
2. "uv add pandas matplotlib jupyterlab notebook": adiciona as bibliotecas matplotlib e pandas no espaço jupyter lab;
3. "uv run jupyter notebook": inicializa um notebook jupyter no navegador;
4. "touch .gitignore": abre o arquivo .gitignore que indica os formatos ou arquivos que o git não deve enviar quando o comando "add ." for acionado.
